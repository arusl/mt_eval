```python
import pandas as pd
import evaluate

df = pd.read_csv('results/results_with_da_scores.tsv', sep='\t')
df
```

```python
refs, preds, srcs = df['Target_En'].tolist(), df['Translation_En_Ours'].tolist(), df['Source_JA'].tolist()

refs2d = []
for ref in refs:
    refs2d.append([ref])

len(refs), len(preds), len(refs2d)
```

```python
# https://huggingface.co/spaces/evaluate-metric/bleu
bleu = evaluate.load("bleu")
bleu_results = bleu.compute(predictions=preds, references=refs2d)
print(bleu_results)
```

```python
# https://huggingface.co/spaces/evaluate-metric/chrf
chrf = evaluate.load("chrf")
chrf_results = chrf.compute(predictions=preds, references=refs2d)
print(chrf_results)
```

```python
# https://huggingface.co/spaces/evaluate-metric/ter
ter = evaluate.load("ter")
ter_results = ter.compute(predictions=preds, references=refs2d)
print(ter_results)
```

```python
# https://huggingface.co/spaces/evaluate-metric/bertscore
bertscore = evaluate.load("bertscore")
berscore_results = bertscore.compute(predictions=preds, references=refs2d, lang="en")
print(berscore_results)
```
