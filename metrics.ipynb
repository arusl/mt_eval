## Evaluate

```python
import evaluate
```

```python
# https://huggingface.co/spaces/evaluate-metric/bleu
bleu = evaluate.load("bleu")
bleu_results = bleu.compute(predictions=preds, references=refs2d)
print(bleu_results)
# {'bleu': 0.1795017974834662, 'precisions': [0.509641873278237, 0.23693534100974314, 0.13164556962025317, 0.07088009450679268], 'brevity_penalty': 0.9797435903021319, 'length_ratio': 0.9799460084843811, 'translation_length': 2541, 'reference_length': 2593}
```

```python
# https://huggingface.co/spaces/evaluate-metric/chrf
chrf = evaluate.load("chrf")
chrf_results = chrf.compute(predictions=preds, references=refs2d)
print(chrf_results)
# {'score': 39.19206275573985, 'char_order': 6, 'word_order': 0, 'beta': 2}
```

```python
# https://huggingface.co/spaces/evaluate-metric/ter
ter = evaluate.load("ter")
ter_results = ter.compute(predictions=preds, references=refs2d)
print(ter_results)
# {'score': 73.89360751005812, 'num_edits': 1653, 'ref_length': 2237.0}
```

```python
# https://huggingface.co/spaces/evaluate-metric/bertscore
bertscore = evaluate.load("bertscore")
berscore_results = bertscore.compute(predictions=preds, references=refs2d, lang="en")
print(berscore_results)
```

## POS Tagging

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
from nltk.tokenize import word_tokenize
prons_ref = []
for text in refs:
    pron_in_ref = []
    for pos in nltk.pos_tag(word_tokenize(text)):
        if 'PRP' in pos or 'PRP$' in pos:
            pron_in_ref.append(pos)
    prons_ref.append(pron_in_ref)
    
prons_pred = []
for text in preds:
    pron_in_pred = []
    for pos in nltk.pos_tag(word_tokenize(text)):
        if 'PRP' in pos or 'PRP$' in pos:
            pron_in_pred.append(pos)
    prons_pred.append(pron_in_pred)
```

```python
print(len(prons_ref))
print(len(prons_pred))
```

```python
for i in range(len(prons_ref)):
    print("Prons in Ref: ", prons_ref[i], " | Prons in Pred: ", prons_pred[i])
```
